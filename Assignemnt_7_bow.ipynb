```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import re
import string
import nltk
pd.set_option('display.max_colwidth', 100)
```

```python

dataset=pd.read_csv(r"F:\Tutorials\Certified program in MACHINE LEARNING and ARTIFICIAL INTELLIGENCE\Assignments\assignment_7\spam.csv")
dataset.head()
```

```python
data=dataset[dataset.columns[0:2]]
data.head()
```

```python
data.columns=['label','body_text']
data.head()
```

```python
df=data
df['label'] = df.label.map({'ham':0, 'spam':1})
print(df.shape)
df.head()
```

```python
stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()
```

```python
def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation]) # remove puctuation
    tokens = re.split('\W+', text)                                                    # tokenize
    text = [ps.stem(word) for word in tokens if word not in stopwords]                #remove stopwords, stemming
    
    return text
```

```python
from sklearn.feature_extraction.text import CountVectorizer                         # BOW using countvector

count_vect = CountVectorizer(analyzer=clean_text)
X_counts = count_vect.fit_transform(data['body_text'])
print(X_counts.shape)
#print(count_vect.get_feature_names())
print(count_vect.get_feature_names_out())
```

```python
X_counts_df = pd.DataFrame(X_counts.toarray())
X_counts_df.columns = count_vect.get_feature_names()
X_counts_df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['body_text'], 
                                                    df['label'], 
                                                    random_state=42,
                                                    test_size=0.3)

print('Total no. of data points: {}'.format(df.shape[0]))
print('No. of data points in training dataset: {}'.format(X_train.shape[0]))
print('No. of data points in testing dataset: {}'.format(X_test.shape[0]))
```

```python
count_vector = CountVectorizer(analyzer=clean_text)

training_data = (count_vector.fit_transform(X_train)).toarray()
print(training_data)
print(training_data.shape)

testing_data = (count_vector.transform(X_test)).toarray()
print(testing_data)
print(testing_data.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators=200)
RF.fit(training_data, y_train)
```

```python
y_RF = RF.predict(testing_data)
```

```python

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print('Accuracy score :' , format(accuracy_score(y_test, y_RF)))
print('Precision score:', format(precision_score(y_test,y_RF)))
print('Recall score:', format(recall_score(y_test, y_RF)))
print('F1 score:', format(f1_score(y_test, y_RF)))
print('\n\n')
```
